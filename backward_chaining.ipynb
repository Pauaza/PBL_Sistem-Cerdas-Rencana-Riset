```python
# Membaca dataset
df = pd.read_csv('master_dataset_dosen.csv')

# Menampilkan data awal
print("Preview Data:")
print(df.head())
```

```python
# Cek nama kolom
print("\nNama Kolom:")
print(df.columns)
```

```python
# Input ide dan deskripsi (GOAL)
ide = input("Masukkan ide skripsi: ").lower()
deskripsi = input("Masukkan deskripsi skripsi: ").lower()

input_user = ide + " " + deskripsi
```

```python
# Fungsi Backward Chaining
def backward_chaining(row, input_user):

    teks = (str(row['judul_penelitian']) + " " +
            str(row['abstrak_penelitian']) + " " +
            str(row['judul_skripsi'])).lower()

    kata_input = input_user.split()
    jumlah_cocok = 0

    # IF 1: cek kecocokan kata
    for kata in kata_input:
        if kata in teks:
            jumlah_cocok += 1

    # IF 2: minimal 2 kata cocok
    if jumlah_cocok >= 2:

        # IF 3: validasi panjang input
        if len(kata_input) >= 3:
            return True, jumlah_cocok

    return False, jumlah_cocok
```

```python
# Proses mencari dosen
hasil = []

for i, row in df.iterrows():

    cocok, skor = backward_chaining(row, input_user)

    if cocok:
        hasil.append({
            'nama_dosen': row['nama_dosen'],
            'lab': row['lab_name'],
            'deskripsi': row['abstrak_penelitian'],
            'skor': skor
        })
```

```python
# Sorting hasil terbaik
hasil = sorted(hasil, key=lambda x: x['skor'], reverse=True)
```

```python
# Fungsi generate judul
def generate_judul(input_user):

    stopwords = ["menggunakan", "untuk", "dan", "yang", "dengan", "berbasis"]

    kata = input_user.split()

    # filter kata
    kata_bersih = []
    for k in kata:
        if k not in stopwords and k not in kata_bersih:
            kata_bersih.append(k)

    kata_utama = " ".join(kata_bersih[:3])

    # hindari double "analisis"
    if "analisis" in kata_utama:
        judul1 = f"{kata_utama.capitalize()} pada Sistem Informasi"
    else:
        judul1 = f"Analisis {kata_utama} pada Sistem Informasi"

    judul2 = f"Perancangan Sistem {kata_utama} Berbasis Web"
    judul3 = f"Implementasi {kata_utama} untuk Meningkatkan Kinerja Sistem"

    return [judul1, judul2, judul3]
```

```python
# Menampilkan hasil
print("\n=== HASIL REKOMENDASI ===")

# tampilkan input user
print("\nInput User:")
print("Ide        :", ide)
print("Deskripsi  :", deskripsi)

# dosen
if len(hasil) > 0:
    print("\nRekomendasi Dosen Pembimbing:\n")

    for h in hasil[:5]:
        print(f"- {h['nama_dosen']} (Lab: {h['lab']}, Skor: {h['skor']})")
else:
    print("Tidak ditemukan dosen yang cocok")


# judul
judul_baru = generate_judul(input_user)

print("\nRekomendasi Judul Skripsi:\n")
for j in judul_baru:
    print("-", j)
```
